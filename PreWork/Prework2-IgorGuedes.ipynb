```python
a = int(input("Please insert a number "))
if a%2 == 0:
  print("It's a even number")
else:
  print("It's a odd number")
if a==0:
  print("The number is 0")
elif a<0 :
  print("The number is negative")
else: 
    print("The number is positive")




```
