```python
import math
print("Print integers from 10 to 30")
for i in range(10,31):
    print(i)
print('Print integers from 10 to 30 in steps of 3')
for i in range(10,31,3):
    print(i)    
print('Print integers from 10 to 30 multiples of 5')
for i in range(10,31):
    if i%5 ==0:
      print(i) 
for i in range(2):
  print('***')
for i in range(10):
  for i in range(1,9,2):
      print(i)
for i in range(2):
  print('****')
n=int(input("Please insert a number :"))
print("Fatorial number for", n ,":",math.factorial(n))

```
