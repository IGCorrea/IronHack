```python
#1.1 - No, if we have a data set with even elements number, we have to make a mean with the central pair numbers
import statistics as stats
data = [4, 7, 5, 9, 8, 6, 7, 7, 8, 5, 6, 9, 8, 5, 8, 7, 4, 7, 3, 6, 8, 9, 7, 6, 9]

mean = stats.mean(data)
print("The mean is: ",mean)
median=stats.median(data)
print("the median is: ",median)
mode=stats.mode(data)
print("the mode is :",mode)
# To find the median we have to put in ascending order
keylist_data =list()
keylist_data.append("mean")
keylist_data.append("median")
keylist_data.append("mode")

valuelist_data=list()
valuelist_data.append(mean)
valuelist_data.append(median)
valuelist_data.append(mode)


data_dictionary=dict()
data_dictionary=dict(zip(keylist_data,valuelist_data))
print(data_dictionary)

#  if we have a data set with even elements number, we have to make a mean with the central pair numbers
#Outliers are points with a big standart deviation from the mean
```
