```python
import statistics as stats
points = [-4, 17, 25, 19, 6, 11, 7]
range = max(points) -min(points)
print(range)
data = [2,4,7,1,6,8]
stdv1=stats.stdev(data)
print(stdv1)
data1=[2,4,7,1,12,8]
stdev2=stats.stdev(data1)
print(stdev2)
# Change 6 for 12 the standart deviation increase
#Yes, is possible if all elements of the data set have the same value
data3=[10,10,10,10,10]
stdev3=stats.stdev(data3)
print(stdev3)

#deviation is the difference between a point from the mean of the data set.Yes, it can be negative
data4 = [23, 12, 34, 65, 34, 81]
mean_data4 =stats.mean(data4)
for i in data4:
  a = mean_data4 - i
  print("The deviation for ", i,"is ",a)
#Unit of variance is squared of the unit of the original data while unit of standard deviation is same as the unit of the original data.
```
