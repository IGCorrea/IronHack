```python
lst = [1,2,34,5,3,12,9, 8, 67, 89, 98, 90, 39, 21, 45, 46, 23, 13]
print("The list have",len(lst),"numbers")
print("The first element on this list is", lst[0])
print("The last element on this list is",lst[len(lst)-1])
print("The position of 90 on this list is",lst.index(90))
print("First 8 numbers on the list",lst[:8])
lst.append(100)
lst.append(110)
print("The actual list:", lst)
print("The actual list have",len(lst),"numbers")
```
