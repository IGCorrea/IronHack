```python
import numpy as np

def posicao_barco_linha1(t,w):
    p2_barco = int(np.random.randint((1),size=1))
    values = np.array([1,1,1])
    
    if p2_barco ==0:
        matriz[(t,t,t),(0,1,2)] = values
        
    else:
        matriz[(0,1,2),(w,w,w)] = values


#criação dos contadores        
print(f'Regras do jogo:\n O jogo consiste em acertar a partir de palpites a localização do barco contido numa matriz 3x3\n Voce deve dar seu palpite no formato (linha, coluna)\n Voce deve inserir a linha que deseja dar seu palpite indexada por a, b ou c\n Voce deve inserir a coluna que deseja dar seu palpite indexada por 1, 2 ou 3\n O jogador pode dar ate 5 coordenadas')       
chutes =0
acertos = 0
size=(3,3)
matriz = np.zeros(size)
 
#posicionamento do barco
p1_barco = int(np.random.randint((9), size = 1))
if p1_barco == 0:
    t=0
    w=0
    posicao_barco_linha1(t,w)
    
if p1_barco == 1:
    t=1
    w=0
    posicao_barco_linha1(t,w)
    
if p1_barco == 2:
    t=2
    w=0
    posicao_barco_linha1(t,w)
    
if p1_barco == 3:
    t=0
    w=1
    posicao_barco_linha1(t,w)
if p1_barco == 4:
    t=1
    w=1
    posicao_barco_linha1(t,w)
if p1_barco == 5:
    t=2
    w=1
    posicao_barco_linha1(t,w)
if p1_barco == 6:
    t=0
    w=2
    posicao_barco_linha1(t,w)
if p1_barco == 7:
    t=1
    w=2
    posicao_barco_linha1(t,w)
if p1_barco == 8:
    t=2
    w=2
    posicao_barco_linha1(t,w)
    
#Entrada de dados
while acertos <= 2 and chutes<=4:
    ent_linha = input('Digite a linha a, b ou c: ')
    ent_col = int(input('Digite a coluna 1, 2 ou 3: '))
    chutes += 1
    if ent_linha =='a':
        if ent_col == 1:
            entrada = np.array([1])
            if entrada == matriz[(0),(0)]:
                acertos+=1
                
        if ent_col == 2:
            entrada = np.array([1])
            if entrada == matriz[(0),(1)]:
                acertos+=1
                
        if ent_col == 3:
            entrada = np.array([1])
            if entrada == matriz[(0),(2)]:
                acertos+=1
                
    if ent_linha =='b':
        if ent_col == 1:
            entrada = np.array([1])
            if entrada == matriz[(1),(0)]:
                acertos+=1
                
        if ent_col == 2:
            entrada = np.array([1])
            if entrada == matriz[(1),(1)]:
                acertos+=1
                
        if ent_col == 3:
            entrada = np.array([1])
            if entrada == matriz[(1),(2)]:
                acertos+=1
               
    if ent_linha =='c':
        if ent_col == 1:
            entrada = np.array([1])
            if entrada == matriz[(2),(0)]:
                acertos+=1
                
        if ent_col == 2:
            entrada = np.array([1])
            if entrada == matriz[(2),(1)]:
                acertos+=1
                
        if ent_col == 3:
            entrada = np.array([1])
            if entrada == matriz[(2),(2)]:
                acertos+=1
    
    
    print(f'Voce acertou: {acertos}')       
    print(f'Voce já tentou {chutes} vezes')
print(matriz)
if acertos ==3:
    print(f'Parabens voce ganhou')
elif chutes == 5:
    print(f'Voce perdeu')
```
